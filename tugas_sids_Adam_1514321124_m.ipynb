```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(1, figsize=(19,13))

# Sample it down to only the Java Island region
lon_min, lon_max = 105.21, 114.60527777778
lat_min, lat_max = -8.78, -5.8758333333333

# Mercator of Java Island
m1 = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max)

m1.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m1.fillcontinents(color='grey', alpha=0.7, lake_color='grey')
m1.drawcoastlines(linewidth=0.1, color="white")

lons = [ 112.6426427,112.5617417,112.5252467,113.166456,113.3732598,114.3949536,110.3470241,110.3398252,108.6952968,107.5731164,106.863809,105.3577406]
lats = [ -7.2756141,-7.9786277,-7.4548864,-7.7723105,-8.2655356,-8.14118,-7.0247246, -7.803249,-7.4616468,-6.9034443,-6.7340184,-6.6188681,]
labels = ['Surabaya', 'Malang', 'Sidoarjo', 'Probolinggo','Jember','Banyuwangi','Semarang','Yogyakarta','Cilacap','Bandung','Cianjur','Pandeglang']

x_offsets = [10000, -30000, -25000]
y_offsets = [5000, -25000, -25000]

x,y = m1(lons, lats)
m1.plot(x, y, 'bo', markersize=10)

plt.title("Pemetaan Kota di Pulau Jawa")

for label, x, y in zip(labels, x, y):
    plt.text(x+10000, y+5000, label)
    
plt.show()
```
